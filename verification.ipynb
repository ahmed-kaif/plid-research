```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('plid.csv')
df = df.dropna()
```

```python
df.columns
```

```python

df = df[['Pre operative ODI', 'Post operative ODI',
       'Pre operative NRS back pain', 'Post operative NRS back pain',
       'Surgery outcome according to Macnab criteria',
       'Pre operative NRS leg pain', 'Post operative NRS leg pain']]

df['mcid'] = df['Pre operative ODI'] - df['Post operative ODI']
df['mcid_leg'] = df['Pre operative NRS leg pain'] - df['Post operative NRS leg pain']
df['mcid_back'] = df['Pre operative NRS back pain'] - df['Post operative NRS back pain']
```

```python
df.head()
```

```python
df.shape
```

```python
def is_mcid_achieved(row):
    if row['Pre operative ODI'] < 20:
        if row['mcid'] >= 2.24: return True
        else: return False
    elif row['Pre operative ODI'] >= 40:
        if row['mcid'] >= 26.57: return True
        else: return False
    elif 20 <= row['Pre operative ODI'] < 40:
        if row['mcid'] >= 11.35 : return True
        else: return False

df['mcid_achieved'] = df.apply(is_mcid_achieved, axis=1)

df.head()
```

```python
import pandas as pd
# Assuming your DataFrame is 'df'

def determine_mcid_outcome(row):
    # Rule 1: mcid not achieved
    if not row['mcid_achieved']:
        return 'P'
    
    # mcid is achieved
    
    # Case 2: Post operative ODI > 30
    elif row['Post operative ODI'] > 30:
        if row['Post operative NRS back pain'] < 2 or row['Post operative NRS leg pain'] < 2:
            return 'G'
        elif row['mcid_leg'] > 2 or row['mcid_back'] > 2:
            return 'F'
        elif row['mcid_leg'] <= 2 or row['mcid_back'] <= 2:
            return 'P' # else P
    
    # Case 3: mcid <= 30
    else: # mcid <= 30
        if row['Post operative NRS back pain'] < 2 or row['Post operative NRS leg pain'] < 2:
            return 'E'
        elif row['mcid_leg'] > 2 or row['mcid_back'] > 2:
            return 'G'
        elif row['mcid_leg'] <= 2 or row['mcid_back'] <= 2:
            return 'F' # else F

# Apply the function row-wise (axis=1)
df['Rule_Outcome_Apply'] = df.apply(determine_mcid_outcome, axis=1)

print(df[['mcid', 'mcid_achieved', 'Rule_Outcome_Apply', 'Surgery outcome according to Macnab criteria']])
```
